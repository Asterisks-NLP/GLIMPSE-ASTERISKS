## Rouge

```python
paths_original = {
    '2017_original': 'output/extractive/rouge/mean/2017-none.csv',
    '2018_original': 'output/extractive/rouge/mean/2018-none.csv',
    '2020_original': 'output/extractive/rouge/mean/2020-none.csv'
}

paths_dynamic = {
    '2017_dynamic': 'output/extractive/rouge/mean/2017-dynamic.csv',
    '2018_dynamic': 'output/extractive/rouge/mean/2018-dynamic.csv',
    '2020_dynamic': 'output/extractive/rouge/mean/2020-dynamic.csv'
}

paths_tfidf = { 
    '2017_tfidf': 'output/extractive/rouge/mean/2017-top.csv',
    '2018_tfidf': 'output/extractive/rouge/mean/2018-top.csv',
    '2020_tfidf': 'output/extractive/rouge/mean/2020-top.csv'
}

paths_clustering = {
    '2017_clustering_1': 'output/extractive/rouge/mean/2017-clustering-1.csv',
    '2017_clustering_2': 'output/extractive/rouge/mean/2017-clustering-3.csv',
    '2017_clustering_3': 'output/extractive/rouge/mean/2017-clustering-5.csv',
    '2017_clustering_5': 'output/extractive/rouge/mean/2017-clustering-7.csv',
    '2017_clustering_7': 'output/extractive/rouge/mean/2017-clustering-10.csv'
}
```

```python
def get_mean_df(paths):
    concat_df = pd.DataFrame()
    for name, path in paths.items():
        df = pd.read_csv(path).drop(columns=['Unnamed: 0']).transpose()
        df.columns = ['rouge-1', 'rouge-2', 'rouge-L']
        df.index = [name]
        concat_df = pd.concat((concat_df, df))
    return (concat_df, concat_df.mean())
```

```python
concat_all_df = pd.DataFrame()
for name, paths in { 'original': paths_original, 'clustering' : paths_clustering, 'tfidf' : paths_tfidf, 'dynamic' : paths_dynamic}.items():
    mean, mean_all = get_mean_df(paths)

    df = pd.DataFrame(mean_all).transpose().set_index(pd.Index([name]))
    concat_all_df = pd.concat((concat_all_df, df))

concat_all_df
```
